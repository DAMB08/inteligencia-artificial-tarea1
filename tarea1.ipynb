```python
import pandas as pd
```

```python
df_json = pd.read_json('csvjson.json')
df_csv = pd.read_csv('cc_state_sector_details.csv')
df = df_csv.merge(df_json, on='stateid',how='left')
```

```python
"""
 Mostrar las ciudades que han  recibido premios a
 nivel estatal en orden descendiente
"""

awards = df.sort_values('awards_per_state_value', ascending=False)
awards[['state_x','awards_per_state_value','year']].drop_duplicates(subset = ["state_x"])
```

```python
"""
 Mostrar las personas se graduaron del género femenino
 en el anio del 2010 de la ciudad de New York en el 
 sistema público
"""
gender = df.gender=='F'
year = df.year==2010
city = df.state_abbr_x == 'NY'
control_x = df.control_x =='Public'
df_gender_year = df[(gender & year) & (city & control_x)]

df_gender_year

```

```python
"""
 Mostrar las personas se graduaron del género masculino
 en el anio del 2012 de la ciudad de Wyoming en el 
 sistema privado
"""
gender = df.gender=='M'
year = df.year==2012
city = df.state_abbr_x == 'WY'
control_x = df.control_x =='Private for-profit'
df_gender_year = df[(gender & year) & (city & control_x)]

df_gender_year
```
